```python
np.random.seed(75243)
temp = notas1.sample(frac=1)['rating']

def calcula_teste(i):
  media = temp[0:i].mean()
  stat, p = ztest(temp[0:i], value=nota_media_por_filme_com_pelo_menos_10_votos.mean()) # 3.4320503405352594
  return (i, media, p)

valores = np.array([calcula_teste(i) for i in range(2, len(temp))])
plt.plot(valores[:,0], valores[:,1])
```

```python
np.random.seed(75243)
temp = notas1.sample(frac=1)['rating']

def calcula_teste(i):
  media = temp[0:i].mean()
  stat, p = ztest(temp[0:i], value=nota_media_por_filme_com_pelo_menos_10_votos.mean()) # 3.4320503405352594
  return (i, media, p)

valores = np.array([calcula_teste(i) for i in range(2, len(temp))])
plt.plot(valores[:,0], valores[:,1])
plt.plot(valores[:,0], valores[:,2])
```

```python
np.random.seed(75243)
temp = notas1.sample(frac=1)['rating']

def calcula_teste(i):
  media = temp[0:i].mean()
  stat, p = ztest(temp[0:i], value=nota_media_por_filme_com_pelo_menos_10_votos.mean()) # 3.4320503405352594
  return (i, media, p)

valores = np.array([calcula_teste(i) for i in range(2, len(temp))])
plt.plot(valores[:,0], valores[:,1])
plt.plot(valores[:,0], valores[:,2])
plt.hlines(y=0.05, xmin=2, xmax=len(temp), colors='red')
```
