### Escolha da Data de Análise

```python
# Pegando a última data do registro
#def lastest_by_country(data):
#  return data.groupby('Country/Region').sum().iloc[:,-1]

# Para testes, será utilizada uma data específica, apresentada em aula
def lastest_by_country(data):
  return data.groupby('Country/Region').sum()['3/2/20']
```

```python
lastest_by_country(confirmed)
```

```python
lastest_by_country(deaths)
```

```python
lastest_by_country(recovered)
```

```python
information = [lastest_by_country(confirmed), lastest_by_country(deaths), lastest_by_country(recovered)]
combined = pd.concat(information, axis=1)
combined.columns = ['confirmed', 'deaths', 'recovered']
combined.head()
```

## Verificando a Taxa de Letalidade

Obs: Expressada em %

```python
combined.sum()
```

```python
sum_up = combined.sum()
```

### Primeira Avaliação

```python
letality_rate_1 = (sum_up['deaths'] / sum_up['confirmed']) * 100
print(letality_rate_1)
```

### Segunda Avaliação

```python
letality_rate_2 = sum_up['deaths'] / (sum_up['recovered'] + sum_up['deaths']) * 100
print(letality_rate_2)
```

### Novas colunas

```python
letality_rate_1 = (combined['deaths'] / combined['confirmed']) * 100
combined['letality_rate_1'] = letality_rate_1
letality_rate_2 = (combined['deaths'] / (combined['recovered'] + combined['deaths'])) * 100
combined['letality_rate_2'] = letality_rate_2
combined.head()
```

```python
combined.sort_values('letality_rate_1', ascending=False).head(10)
```

```python
combined.sort_values('letality_rate_2', ascending=False).head(10)
```
